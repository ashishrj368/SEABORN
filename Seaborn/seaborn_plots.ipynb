```python
# scatter plot

import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create a histogram of total bill amounts
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
# Barplot

sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
plt.show
```

```python
# BOX PLOT

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.pairplot(tips, hue='tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.pairplot(tips)
plt.show()
```

```python
# DISPLOT
# creating with kernel density estimate plot

sns.displot(tips['total_bill'], kind='kde')
plt.show()
```

```python
# DISPLOT
# creating with histogram estimate plot

sns.displot(tips, x='total_bill', hue='sex', kind='hist', bins=30)
plt.show()
```
